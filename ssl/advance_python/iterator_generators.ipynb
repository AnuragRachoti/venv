```python
##iterators --> usually used for efficient looping and powerfull memory management.
##when we print iterator it doesnt give us the values it gives us the address
## so basically it uses lazy loading technique.

li = [1,2,3,4,5,6]

iterator = iter(li)
print(iterator)
```

```python
##its not gonna give the value of particular element untill and unless it iterates the order of particular element
## to iterate we use another function called next
try:
    print(next(iterator))
except StopIteration:
    print('end of iteration values')
```

###generators
generators are a simpler way to create iterators.

They use the yield keyword to produce a series of values lazily, which means they generate values on the fly and donot store them in memory.

```python
def square(n):
    for i in range(3):
        yield i**2

for i in square(3):
    print(i)
```

```python
a = square(3)
next(a)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
```

```python
gen = (my_generator())
```

```python
next(gen)
```

```python
for val in gen:
    print(val)
```

###practical example : reading large files
### generators are particulary useful for reading large files because they allow you to process one line at a time without loading the entire file into memory.

```python
def read_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line
```

```python
file_path = 'largefile.txt'

for line in read_file(file_path):
    print(line.strip())
```
