```python
##filling missing values with the mean of the column
##usually we use the below operation to handle the null values as the entire column has NAN values for time being im replacing it with 0
##df['Response_fillNa'] = df['Response'].fillna(df['Response'].mean())

df = df.fillna(0)
df
```

```python
grouped_mean = df.groupby('Class')['LocationID'].mean()
```

```python
grouped_mean
```

```python
##aggregating multiple functions.

grouped_agg = df.groupby('Class')['YearEnd'].agg(['mean','sum','count'])
```

```python
grouped_agg
```

```python
###merging and joining dataframes

df1 = pd.DataFrame({'Key' : ['A','B','C'], 'Value1' : [1,2,3]})
df2 = pd.DataFrame({'Key' : ['A','B','D'], 'Value2' : [4,5,6]})
```

```python
df1
```

```python
df2
```

```python
###merge dataframe on the 'Key columns'ArithmeticError
df_inner = pd.merge(df1,df2, on='Key',how='inner')
df_inner
```

```python
df_outer= pd.merge(df1,df2, on='Key',how='outer')
df_outer
```

```python
df_left= pd.merge(df1,df2, on='Key',how='left')
df_left
```

```python
df_right= pd.merge(df1,df2, on='Key',how='right')
df_right
```
